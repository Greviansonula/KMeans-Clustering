```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
# Load the Iris dataset
iris = load_iris()
iris_df = iris.data
```

```python
# Perform k-means clustering for different values of k
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=5)
    kmeans.fit(iris_df)
    sse[k] = kmeans.inertia_ # Inertia: Sum of distances of samples to their closest cluster center
```

```python
# Plot the elbow curve
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.show()
```

```python
# Get the cluster labels
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(iris_df)
centroids = kmeans.cluster_centers_
```

```python
# Plot the data before clustering
plt.scatter(iris_df[:, 0], iris_df[:, 1])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset (before clustering)')
plt.show()
```

```python
# Get the cluster labels
labels = kmeans.predict(iris_df)
```

```python
# Plot the data after clustering
colors = ['r', 'g', 'b']
for i in range(3):
    plt.scatter(iris_df[labels == i, 0], iris_df[labels == i, 1], c=colors[i])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='black', s=300)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset (after clustering)')
plt.show()
```
